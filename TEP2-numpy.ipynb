```python
import numpy as np
import matplotlib.pyplot as plt
imagem = plt.imread('arvore.png')

print('-' *50)
print(f'Shape do array da imagem: {imagem.shape}')
print(f'Tipo de dados da imagem: {imagem.dtype}')
print(f'Dimensões da imagem: {imagem.ndim}')
print('-' *50)
```

```python
plt.figure(figsize=(8,4))
plt.imshow(imagem)
plt.title('Imagem original')
plt.axis('off')
plt.show()
```

```python
print(imagem[100, 100])
```

```python
cinza = np.mean(imagem, axis=2)
print(cinza.shape)

plt.figure(figsize=(8,4))
plt.imshow(cinza, cmap='gray')
plt.title('Imagem preto e branco')
plt.axis('off')
plt.show()
```

```python
altura, largura = imagem[:2]
#y1, y2 = altura // 4, 3 * altura // 4
#x1, x2 = largura // 4, 3 * largura // 4

recorte = imagem[100:250, 100:250]
print(recorte.shape)

cinza = np.mean(imagem, axis=2)
print(cinza.shape)

plt.figure(figsize=(8,4))
plt.imshow(recorte)
plt.title('Imagem recortada')
plt.axis('off')
plt.show()
```

```python
imagemb = imagem.copy()
imagemb[:,:, 0] = 0

plt.figure(figsize=(8,4))
plt.imshow(imagemb)
plt.title('Imagem sem a cor vermelha')
plt.axis('off')
plt.show()
```

```python
print(imagem.max())
print(imagem.min())
```

```python
if imagem.max() <= 1.0:
    imagem_inv = 1.0 - imagem
else:
    imagem_inv = 255 - imagem

plt.figure(figsize=(8,4))
plt.imshow(imagem_inv)
plt.title('Imagem contorno')
plt.axis('off')
plt.show()
```

```python
imagem_esph = imagem[:, ::-1]

plt.figure(figsize=(8,4))
plt.imshow(imagem_esph)
plt.title('Imagem espelhada')
plt.axis('off')
plt.show()
```

```python
imagem_invertida = imagem[::-1]

plt.figure(figsize=(8,4))
plt.imshow(imagem_invertida)
plt.title('Imagem invertida')
plt.axis('off')
plt.show()
```

```python
imagem_90 = np.rot90(imagem, k=1)

plt.figure(figsize=(8,4))
plt.imshow(imagem_90)
plt.title('Imagem virada a 90 graus')
plt.axis('off')
plt.show()
```
